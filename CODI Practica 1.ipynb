Per a aquest primer apartat es calcularà l'error quadràtic mitjà (MSE) només del regressor per a cada un dels atributs de la base de dades, determinant aquell atribut pel qual el MSE (entre el valor predit i el real, per a cada mostra) és més baix. 

A continuació se us dona una funció auxiliar per a calcular l'error quadràtic mitjà:

Per a la regressió podeu utilitzar la llibreria sklearn:

````python
from sklearn.linear_model import LinearRegression

def regression(x, y):
    # Creem un objecte de regressió de sklearn
    regr = LinearRegression()

    # Entrenem el model per a predir y a partir de x
    regr.fit(x, y)

    # Retornem el model entrenat
    return regr
````

Càlcul dels MSE i rectes de regressió per cada atribut

````python
# es pot utilitzar numpy per a calcular el mse
def mse(v1, v2):
    return ((v1 - v2)**2).mean()
````

````python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
regr = LinearRegression()
#mejores 6, 7 ->  19, 15
#peor el 1 y 9 -> 60, 62


RSE_vector = np.array([])
R2_vector = np.array([])
fig, axes = plt.subplots(4, 4, figsize=(18, 15))

fig.suptitle("REGRESIONS LINEALS PER CADA ATRIBUT")
for i in range(x_sense.shape[1]):
    eix_x = int(i/4)
    eix_y = i%4

    xi = x_sense[:,i]
    xi = xi.reshape(-1,1)
    regr.fit(xi, y)
    pred_xi = regr.predict(xi)
    RSE_vector = np.append(RSE_vector, metrics.mean_squared_error(pred_xi, y))
    print ("MSE atribut ",x_labels[i],": ", int(metrics.mean_squared_error(pred_xi, y)))
    R2_vector = np.append(R2_vector, metrics.r2_score(pred_xi, y))
    print ("R^2: atribut ",x_labels[i],": ", metrics.r2_score(pred_xi, y))
    print("")
    
    axes[eix_x, eix_y].plot(xi, y,'o', alpha = 0.5)
    axes[eix_x, eix_y].plot(xi, pred_xi, 'r', alpha = 0.5)
    axes[eix_x, eix_y].set_title(str(x_labels[i]))

````

# Apartat (A): El descens del gradient  

En aquest exercici, es tracta d'implementar en python el procés de descent del gradient explicat a les classes de teoria, i comparar-lo amb els resultats obtinguts amb l'apartat (B). 

$$J(w) = \frac{1}{2m} \left[ \sum^m_{i=1}(f(x^{i}; w) - y^{i})^2 + \lambda\sum_{j=1}^{n}(w_{j}^2) \right]$$

Fixeu-vos que $J$ retorna el `mse`. Per a trobar $w_j$, repetir fins convergència:
$$w_0 = w_0 - \alpha \frac{1}{m} \sum_{i=1}^{m}(f(x^{i}; w)-y^{i}) \cdot 1$$
$$w_j = w_j - \alpha \left[\frac{1}{m} \sum_{i=1}^{m}(f(x^{i}; w)-y^{i}) \cdot x_{j}^{i} - \frac{\lambda}{m}w_{j} \right]$$


ó:

$$w_{j} := w_{j} \left(1-\alpha \frac{\lambda}{m} \right) - \alpha\frac{\lambda}{m} \sum_{i=1}^{m}(f(x^{i}; w)-y^{i}) \cdot x_{j}^{i}$$

On si considerem un regressor lineal (el model és una recta), llavors $w_0$ i $w_1$ representen, respectivament, la $b$ i $a$ de la fòrmula de la recta: 

$$h_\theta(x^{(i)}) = ax + b$$

$\alpha$ és el learning rate, i $h_\theta(x^{(i)})$ és la funció que fa la regressió, és a dir, la funció que prediu el valor de $y^{(i)}$ donat un(s) atribut(s) concret(s) $x^{(i)}$.

Així, tenint calculat el model en l'últim punt del primer exercici, ja sabeu quin resultat hauríeu d'obtenir. O no, perquè la vostra implementació pot ser millor! En concret, es tracta de desenvolupar aquestes tasques:

Per a la implementació us podeu basar en el següent esquelet:

````python
from math import inf


class Regressor(object):
    def __init__(self, x, alpha, landa=0):
        # Inicialitzem totes les Ws que x permeteixi (num de columnes)
        self.w = []
        self.alpha = alpha
        self.landa = landa
        
        print("Init values: ", end="")

        for i, column in enumerate(x.T):
          self.w.append(column[0])
          print(f"W{i}: {column[0]} ", end="")
          
        print()
        self.w = np.array(self.w)

        
    def predict(self, x):
        # predicció --> h(x) = ax^0 + bx^1 + cx^2 + dx^3...

        final_result = 0

        for i, wi in enumerate(self.w):
          final_result += wi * (x[:,i] ** i)

        return final_result
    
    def __update(self, hy, y, x):
        # actualitzar aqui els pesos donada la prediccio (hy) i la y real.

        # m seran els elements totals
        m = hy.size

        for i, wi in enumerate(self.w):
          # perdua entre hipotesis i valor real
          loss = hy - y

          # multiplicació de la perdua per totes les xij
          inner_sum = np.inner(x[:,i] ** i, loss)

          # multiplicació de lambda per el wi actual
          landa_const = self.alpha * (self.landa / m)

          # fórmula per obtenir la w actual
          self.w[i] = wi * (1 - landa_const) - landa_const * inner_sum

        landa_cost = self.landa * np.sum(self.w ** 2)
        total_cost = (np.sum(loss ** 2) + landa_cost) / (2 * m)
        
        return total_cost
        
    def train(self, max_iter, epsilon, x, y):
        # Entrenar durant max_iter iteracions o fins que la millora sigui inferior a epsilon
        last_cost = float(inf)
        y = y.flatten()

        for i in range(max_iter):
          # print(f"Iteration {i} | W0 = {self.w[0]}, W1 = {self.w[1]}")

          hy = self.predict(x)
          cost = self.__update(hy, y, x)

          # print(f"Actual cost J of iteration {i} is {cost:_}")

          if (last_cost - cost < epsilon): break

          last_cost = cost

        print(f"Total iterations: {i}")
        print(f"Best cost is: {cost:_}")
        print(f"Best Ws were: ", end="")
        for i, wi in enumerate(self.w): print(f"W{i}: {wi} ", end="")
        print()
````
